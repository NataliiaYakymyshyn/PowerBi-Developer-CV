```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.keys())

```

Опис словами (треба виділити головне)

```python
print(iris.DESCR)
```

Набір даних Iris - це класичний набір даних у сфері машинного навчання, який часто використовується для демонстрації алгоритмів класифікації та кластеризації. Цей набір даних описує характеристики трьох класів рослин Iris.

**Ознаки (атрибути) набору даних Iris:**

1. **Довжина чашелистника (sepal length).** Вимірюється в сантиметрах та представляє довжину першого чашелистника квітки.

2. **Ширина чашелистника (sepal width).** Вимірюється в сантиметрах та представляє ширину першого чашелистника квітки.

3. **Довжина пелюстки (petal length).** Вимірюється в сантиметрах та представляє довжину першої пелюстки квітки.

4. **Ширина пелюстки (petal width).** Вимірюється в сантиметрах та представляє ширину першої пелюстки квітки.

**Клас (цільова змінна):**

Кожен запис у наборі даних Iris відноситься до одного з трьох видів ірису:

1. **Iris-setosa**
2. **Iris-versicolor**
3. **Iris-virginica**

Кожен вид ірису має свої характерні особливості щодо довжини та ширини чашелистника та пелюстки, що дозволяє використовувати цей набір даних для вивчення та демонстрації алгоритмів класифікації в машинному навчанні.

```python
print(iris.target)
```

```python
df_iris = pd.DataFrame(iris['data'], columns = iris['feature_names'])
df_iris.head()
```

Перевірка наявностi NA значень

```python
df_iris.info()
```

Вивід базових статистик

```python
df_iris.describe()
```

```python
df_target_iris = pd.DataFrame(iris['target'], columns =[ 'species'])
df_iris_all = pd.concat([df_iris, df_target_iris], axis=1)
```

```python
df_iris_all
```

Розподіл кількості спостережень за класами

Гістограма

```python
sns.countplot(data=df_iris_all, x='species')
```

Кругова діаграма

```python
count_list = [(df_iris_all.species == 0).sum(), (df_iris_all.species == 1).sum(), (df_iris_all.species == 2).sum()]
print(count_list)
label_list = list(df_iris_all['species'].unique())
print(label_list)
plt.figure(figsize = (10, 7))
plt.pie(count_list, labels = label_list, autopct = "%.2f %%", startangle = 90, explode = (0.1, 0.1, 0.0), textprops = {'fontsize': 12})
plt.title('Distribution of Classes', fontsize = 20)
plt.show()
```

Розподіл ознак, гістограма

```python
df_iris_all.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(15,6)
plt.show()
```

Точкова діаграма (scatterplot) розподілу спостережень за класами

```python
sns.pairplot(df_iris_all, hue='species', palette= "tab10", diag_kind='hist')
```

Кореляційна матриця

```python
plt.figure(figsize=(7,4))
sns.heatmap(df_iris_all.corr(),annot=True,fmt="f",cmap="YlGnBu")
plt.show()
```

**Висновок.** Базуючись на точковій діаграмі розподілу спостережень за класами та кореляційній матриці, можна зробити висновок, що найбільш перспективними з точки зору подальшої класифікації ознаками є

- petal width
- petal length

На точковій діаграмі видно, що саме за ціма двома ознаками спостереження різних класів (точки різних кольорів) добре розділаються. Також цей висновок підтверджується значеннями кореляції цих ознак з цільовою змінною (0.96, 0.95 відповідно)
